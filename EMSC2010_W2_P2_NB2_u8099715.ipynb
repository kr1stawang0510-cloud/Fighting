# Fermi estimation problem: How many water droplets fall in a storm

### This code was written by Jet, Lee, Binyao and Kedi

```python
raindrop_size = 0.1 #in ml

# 1ml = 1 cm3
# 1ml = 1000 mm3

raindrop_mm3 = 1000 * raindrop_size
```

```python
mm_of_rainfall = 50 #in mm
```

```python
area_of_storm = 5 #km^2
```

```python
area_in_mm = area_of_storm * (1000**2 * 100**2 * 10**2) #from km2 to m2 to cm2 to mm2
```

Standardized the volume conversion from km2 to mm2.

```python
volume = area_in_mm*mm_of_rainfall
```

```python
raindrops = volume / raindrop_mm3
```

```python
print(raindrops)
```

```python
print(f"There are {raindrops:.1e} raindrops in a {area_of_storm} km^2 storm.")
```
